```python
print(df.columns)
df = df[['AcresBurned','Active','ArchiveYear','CanonicalUrl','Counties','CountyIds','Extinguished','Latitude','Location','Longitude','Name','Started']]
```

```python
# Returns a list of zipcodes with a radious
# radius is default zero
def get_zips(df, lat,lng,r=30):
    search = SearchEngine()
    try:
        results = search.by_coordinates(lat=lat,lng=lng, radius=r)
        return [r.zipcode for r in results]
    except (AttributeError, KeyError, ValueError):
        return None
```

```python
df['zipcode'] = df.progress_apply(get_zips, axis=1, lat = df['Latitude'], lng = df['Longitude'])
```
